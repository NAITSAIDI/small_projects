# Create a nice little circle using turtle

```python
for colours in ['red', 'yellow', 'green', 'blue', 'purple', 'pink']:
    turtle.color(colours)
    turtle.circle(150)
    turtle.left(60)
    turtle.speed(10)
turtle.done()    
```
